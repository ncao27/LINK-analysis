```python
# IMPORT NECESSARY LIBRARIES

from pynwb import NWBHDF5IO as nwb
import pynapple as nap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import nemos as nmo
import os
from nemos import _documentation_utils as doc_plots
```

```python
# IMPORT RECORDINGS

folder_path = "../data/dandi/001201/sub-Monkey-N/"

# List all NWB files recursively
nwb_files = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith(".nwb"):
            nwb_files.append(os.path.join(root, file))

print(f"Found {len(nwb_files)} NWB files.")

# Open the first file
file_path = nwb_files[0]
print("\nOpening:", file_path)

io = nwb(file_path, "r")
nwbfile = io.read()
print(nwbfile)
```

```python
print("========== ANALYSIS KEYS ==========")
print(list(nwbfile.analysis.keys()))

print("\n========== TRIALS ==========")
trials_df = nwbfile.trials.to_dataframe()
print("Trials shape:", trials_df.shape)
print("Trials columns:", trials_df.columns.tolist())
print(trials_df.head())

print("\n========== ELECTRODES ==========")
electrodes_df = nwbfile.electrodes.to_dataframe()
print("Electrodes shape:", electrodes_df.shape)
print("Electrodes columns:", electrodes_df.columns.tolist())
print(electrodes_df.head())

print("\n========== ANALYSIS OBJECT DETAILS ==========")
for key, obj in nwbfile.analysis.items():
    print(f"\n--- {key} ---")
    print("Type:", type(obj))
    
    if hasattr(obj, "description"):
        print("Description:", obj.description)
    if hasattr(obj, "unit"):
        print("Unit:", obj.unit)
    if hasattr(obj, "rate"):
        print("Rate:", obj.rate)
    if hasattr(obj, "starting_time"):
        print("Starting time:", obj.starting_time)
    
    # shape of data
    if hasattr(obj, "data"):
        try:
            print("Data shape:", obj.data.shape)
        except Exception as e:
            print("Could not get data shape:", e)
    
    # timestamps info
    if hasattr(obj, "timestamps") and obj.timestamps is not None:
        try:
            print("Num timestamps:", len(obj.timestamps))
            print("First 5 timestamps:", obj.timestamps[:5])
        except Exception as e:
            print("Could not inspect timestamps:", e)
```

```python
print("\n========== SAMPLE DATA PREVIEW ==========")

for key in nwbfile.analysis.keys():
    obj = nwbfile.analysis[key]
    print(f"\n--- {key} ---")
    try:
        data = obj.data[:10]
        print("First few samples shape:", np.array(data).shape)
        print(data)
    except Exception as e:
        print("Could not preview data:", e)
```

```python
import matplotlib.pyplot as plt

behavior_keys = ["index_position", "index_velocity", "mrs_position", "mrs_velocity"]

for key in behavior_keys:
    if key in nwbfile.analysis:
        obj = nwbfile.analysis[key]
        data = np.asarray(obj.data[:2000])

        plt.figure(figsize=(10, 3))
        if data.ndim == 1:
            plt.plot(data)
        else:
            plt.plot(data[:, 0])
        plt.title(key)
        plt.xlabel("Sample")
        plt.ylabel("Value")
        plt.show()
```
